# Cohort Analysis on E-Commerce Data
Macam-macam cohort analysis:
1. **Acquisition Cohort** - Pengelompokkan pelanggan berdasarkan kapan pelanggan bergabung/bertransaksi pertama kalinya. Ada beberapa hal yang bisa kita pantau perubahannya seiring berjalannya waktu (misal: per hari, bulan, atau tahun), yaitu:
    * Perubahan jumlah pelanggan
    * Perubahan jumlah spending
2. 

Secara garis besar, terdapat 3 tahapan yang perlu kita lakukan untuk melakukan cohort analysis, diantaranya:
1. Mempersiapkan kolom-kolom yang dibutuhkan
2. Membuat tabel retensi
3. Membuat visualisasi dalam bentuk heatmap

Ada 3 kolom yang perlu dibuat agar kita dapat melakukan cohort analysis, yaitu:
1. Invoice periode - waktu (hari, bulan, atau tahun) terjadinya sebuah transaksi
2. Cohort group - cohorts (kelompok pelanggan) berupa waktu (hari, bulan, atau tahun) kapan pelanggan melakukan transaksi pertamanya
3. Cohort period - periode waktu (hari, bulan, atau tahun) yang digunakan untuk mencatat perkembangan perilaku pelanggan setiap waktunya

#### Data preparation before doing the analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
# read dataset
df = pd.read_csv('../datasets/cleaned_data.csv')
```

```python
# delete 'Unnamed: 0' column
del df['Unnamed: 0']
```

```python
# get data dimension
print('Number of columns:', df.shape[0])
print('Number of rows:', df.shape[1])
```

```python
# change the format of 'InvoiceDate' to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

## 1. Prepare some columns needed for analysis
Ada 3 kolom yang perlu dibuat agar kita dapat melakukan cohort analysis, yaitu:
1. Invoice periode - waktu (hari, bulan, atau tahun) terjadinya sebuah transaksi
2. Cohort group - cohorts (kelompok pelanggan) berupa waktu (hari, bulan, atau tahun) kapan pelanggan melakukan transaksi pertamanya
3. Cohort period - periode waktu (hari, bulan, atau tahun) yang digunakan untuk mencatat perkembangan perilaku pelanggan setiap waktunya

#### Task 1. Prepare a column for the invoice period
Mengingat nantinya kita akan membuat kolom cohorts (kelompok pelanggan), maka kita memerlukan periode invoice yang nilai terkecilnya (bulan paling awal) akan digunakan sebagai nilai cohorts. Oleh sebab itu, kita akan mengambil nilai tanggal tanpa jam, menit, detik untuk mengisi kolom 'InvoicePeriod'.

Mengingat bahwa data ini mencatat transaksi yang terjadi selama 13 bulan, maka nantinya akan terbentuk cohorts (pengelompokan pelanggan) dengan satuan bulan.

```python
# create new column 'InvoicePeriod' extracted from 'InvoiceDate'
df['InvoicePeriod'] = df['InvoiceDate'].map(lambda d: dt.datetime(d.year, d.month, 1))
```

```python
# check the first 5 rows of the data
df.head()
```

#### Task 2. Create the cohorts group

```python
# create a new column 'Cohorts' extracted from the earliest InvoicePeriod
df['Cohorts'] = df.groupby('CustomerID')['InvoicePeriod'].transform('min')
```

```python
df.head()
```

#### Task 3. Prepare the cohort periods
Dataset ini merupakan data transaksi yang terjadi selama 13 bulan. Pada analisis kali ini kita akan memantau perubahan perilaku pelanggan setiap bulannya, dari bulan ke-0 (Month 0) hingga terakhir (Month 12). Setiap transaksi/invoice akan memiliki ditandai dengan angka yang menunjukkan apakah transaksi tersebut terjadi pada bulan ke-0 (Dec '10), ke-1 (Jan '11), dst.

Kita bisa mendapatkan angka tersebut dengan mengambil selisih antara waktu tercatatnya setiap transaksi (InvoicePeriod) dengan waktu tercatatnya transaksi pertama dari pelanggan terkait (Cohorts).

```python
# get the year & month of transaction date as integer
def get_year_month(df, col):
    year = df[col].dt.year
    month = df[col].dt.month
    return year, month
```

```python
# get the year & month values for invoice & cohort dates as integer
invoice_year, invoice_month = get_year_month(df, 'InvoicePeriod')
cohort_year, cohort_month = get_year_month(df, 'Cohorts')
```

```python
# get the difference of year & month values of the invoice & cohort dates
year_diff = invoice_year - cohort_year
month_diff = invoice_month - cohort_month
```

```python
# create a new column 'CohortPeriod' extracted from the difference between invoice and cohort dates
df['CohortPeriod'] = year_diff * 12 + month_diff
```

```python
df.head()
```

Kita telah membuat tiga kolom baru yang diperlukan: InvoicePeriod, Cohorts, dan CohortPeriod.

Langkah selanjutnya yang perlu kita lakukan adalah membuat tabel retensi.

## 2. Create the Retention Table

```python
# prepare a grouping of dataframe by 'Cohorts' and 'CohortPeriod' columns values
grouping = df.groupby(['Cohorts', 'CohortPeriod'])
```

```python
# get the number of customer for each combination of cohort groups and cohort periods
customer_per_cohort_group_and_period = grouping['CustomerID'].apply(pd.Series.nunique).reset_index()
customer_per_cohort_group_and_period
```

```python
# create a retention table containing the number of active customer for each cohort periods
customer_number_cohort_table =  customer_per_cohort_group_and_period.pivot(index='Cohorts', columns='CohortPeriod', values='CustomerID')
customer_number_cohort_table
```

```python
# get the total number of customer in each cohorts
cohort_sizes = customer_number_cohort_table.iloc[:,0]
cohort_sizes
```

Based on the output above, we know that the first cohort (Dec '10) has 885 customers, the second cohort (Jan '11) has 417 customers, etc. It means that there were 885 customers making the first transaction on Dec '10 and 417 customers making the first transaction on Jan '11.

```python
# get the retention rate (%) of customer who did a transaction again for each months
retention = customer_number_cohort_table.divide(cohort_sizes, axis=0) * 100

# round the value on 1 floating point
retention = retention.round(1)

# redefine the retention table with the percentage of customer that still active in each cohort period based on each cohort sizes
retention_table = pd.merge(customer_number_cohort_table.iloc[:,0], retention, on="Cohorts")
retention_table
```

```python
# rename the first two columns
retention_table.rename(columns={'0_x': '#customers', '0_y': '0'}, inplace=True)

# reset the data type of the first two columns
retention_table['#customers'] = retention_table['#customers'].astype('int')
retention_table.iloc[:,1] = retention_table.iloc[:,1].astype('int') 

retention_table
```

```python
# prepare more readable labels for each cohorts
cohort_labels = ["Dec '10", "Jan '11", "Feb '11", "Mar '11", "Apr '11", \
                 "May '11", "Jun '11", "Jul '11", "Aug '11", "Sep '11", \
                 "Oct '11", "Nov '11", "Dec '11"]

# create a temporary dataframe for the cohort labels
cohorts = pd.DataFrame({'cohort': cohort_labels}, index=retention_table.index)

# merge the retention_table with the temporary dataframe above (cohorts)
retention_table = pd.merge(cohorts, retention_table, on='Cohorts')

# set a new index for retention table
retention_table.set_index('cohort', inplace=True)

retention_table
```

## 3. Visualize the Retention Table

```python
# set the width and height of the figure
plt.figure(figsize=(14,7))

# set the title of the heatmap
plt.title('Monthly Retention Rates')

# setup the heatmap
sns.heatmap(data=retention_table.iloc[:,1:14], 
            annot=True, 
            fmt='.1f', 
            cmap='BuGn',
            vmax=list(retention_table.max().sort_values(ascending=False))[2])

plt.show()
```
